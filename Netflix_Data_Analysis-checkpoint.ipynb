### Identifying and Treating NULL/ NaN Values

```python
netflix_data_copy= netflix_data.copy()

netflix_data_copy
```

```python
# Removing unnecessary columns
```

```python
netflix_data_copy.drop(['director','cast','date_added'], axis=1, inplace= True)

netflix_data_copy
```

```python
netflix_data_copy.isnull()
```

```python
netflix_data_copy.isnull().sum()
```

```python
netflix_data_copy['country'].value_counts()
```

```python
netflix_data_copy['country'].replace(np.NaN, 'United States', inplace=True)

netflix_data_copy
```

```python
netflix_data_copy.head(5)
```

```python
netflix_data_copy['rating'].value_counts()
```

```python
netflix_data_copy['rating'].replace(np.NaN,'TV-MA', inplace= True)

netflix_data_copy
```

```python
netflix_data_copy['duration'].value_counts()
```

```python
netflix_data_copy[netflix_data_copy['duration'].isnull()]
```

```python
netflix_data_copy.isnull().sum()
```
